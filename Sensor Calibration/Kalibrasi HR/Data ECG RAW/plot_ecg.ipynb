```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# ====== PARAMETER ======
file_path = 'ecg_raw_125hz_wawan2.csv'  # Ubah jika lokasi file berbeda
sampling_rate = 125  # Hz
start_second = 10
end_second = 20
```

```python
# ====== LOAD DATA ======
data = pd.read_csv(file_path)
ecg = data['ecg_value']
```

```python
# ====== PLOT SELURUH SINYAL ======
plt.figure(figsize=(14, 4))
plt.plot(ecg, color='orange')
plt.title('Sinyal ECG Lengkap - 125 Hz')
plt.xlabel('Nomor Sampel')
plt.ylabel('Amplitudo ECG (ADC Unit)')
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# ====== PLOT CUPLIK DETIK 10-20 ======
start_idx = start_second * sampling_rate
end_idx = end_second * sampling_rate

plt.figure(figsize=(14, 4))
plt.plot(ecg[start_idx:end_idx], color='orange')
plt.title(f'Cuplikan Sinyal ECG Detik {start_second}-{end_second} (125 Hz)')
plt.xlabel('Nomor Sampel')
plt.ylabel('Amplitudo ECG (ADC Unit)')
plt.grid()
plt.tight_layout()
plt.show()
```
